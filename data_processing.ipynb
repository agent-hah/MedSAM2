```python
import os
import numpy as np
from tqdm import tqdm

def calculate_global_stats(data_dir):
    print(f"Calculating global mean and std for dataset: {data_dir}")

    files = [f for f in os.listdir(data_dir) if f.endswith('.npz')]

    pixel_sum = 0.0
    pixel_sq_sum = 0.0
    total_pixels = 0

    for f in tqdm(files, desc="Processing Volumes"):
        data = np.load(os.path.join(data_dir, f))
        # Load the image volume and scale to [0, 1]
        imgs = data['imgs'].astype(np.float64) / 255.0

        pixel_sum += np.sum(imgs)
        pixel_sq_sum += np.sum(imgs ** 2)
        total_pixels += imgs.size

    global_mean = pixel_sum / total_pixels
    # Variance = E[X^2] - (E[X])^2
    global_var = (pixel_sq_sum / total_pixels) - (global_mean ** 2)
    global_std = np.sqrt(global_var)

    print("\n" + "="*45)
    print("      DATASET NORMALIZATION STATS      ")
    print("="*45)
    print(f" Mean : {global_mean:.4f}")
    print(f" Std  : {global_std:.4f}")
    print("="*45)
    print("\nFormat for MedSAM2 (copy these):")
    print(f"image_mean: [{global_mean:.4f}, {global_mean:.4f}, {global_mean:.4f}]")
    print(f"image_std:  [{global_std:.4f}, {global_std:.4f}, {global_std:.4f}]")

if __name__ == "__main__":
    # Point this to your preprocessed DIAS_Train_NPZ folder
    calculate_global_stats("data/medsam_preprocessed/DIAS_Standard_NPZ")
```
